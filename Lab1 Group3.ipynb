```
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import numpy as np
```

###Question 31 from Chapter 12.6

$z=\sqrt{9y^2-9x^2}$

```
def f(x, y):
  return np.sqrt((9*Y**2 - 9*X**2))

x = np.linspace(-1, 1)
y = np.linspace(-1, 1)

X, Y = np.meshgrid(x, y)
Z = np.sqrt((9*Y**2 - 9*X**2))

ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z)
ax.set_title('$z = \sqrt{(9y^2 - 9x^2)}$')
ax.set_xlabel('x')
ax.set_ylabel('y')
```

###Question 9 from Chapter 14.1

$g(x,y)=cox(x+2y)$

```
def g(x, y):
  return np.cos(x + 2*y)

x = np.linspace(-1, 1)
y = np.linspace(-1, 1)

X, Y = np.meshgrid(x, y)
Z = g(X, Y)

ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z)
ax.set_title('$g(x,y)=cox(x+2y)$')
ax.set_xlabel('x')
ax.set_ylabel('y')
```

###Question 10 from Chapter 14.1

$F(x,y)=1+\sqrt{(4-y^2)}$

```
def F(x, y):
  return 1 + (4 - y**2)**(1/2)

x = np.linspace(-1, 1)
y = np.linspace(-1, 1)

X, Y = np.meshgrid(x, y)
Z = F(X, Y)

ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z)
ax.set_title('$F(x,y)=1+\sqrt{(4-y^2)}$')
ax.set_xlabel('x')
ax.set_ylabel('y')
```

###Question 32a from Chapter 14.1
